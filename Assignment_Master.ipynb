```python
import pandas as pd
```

```python
def convert_txt_to_csv(input_file, output_file, delimiter):
  df = pd.read_csv(input_file, delimiter=delimiter)
  df.to_csv(output_file, index=False)
```

```python
def json_to_csv(input_file, output_file):
  df = pd.read_json(input_file)
  df.to_csv(output_file, index=False)
```

```python

# Load the CSV files to inspect their structure
angelone_df = pd.read_csv('/content/angelone.csv')
finvasia_df = pd.read_csv('/content/finvasia.csv')
aliceblue_df = pd.read_csv('/content/aliceblue.csv')
flattrade_df = pd.read_csv('/content/flattrade.csv')
zerodha_df = pd.read_csv('/content/zerodha.csv')

# Display the first few rows of each DataFrame to understand their structure
angelone_df.head(), finvasia_df.head(), flattrade_df.head(), zerodha_df.head(), aliceblue_df.head()

```

```python
# Define the desired column order
desired_order = ['token', 'name', 'trading_symbol', 'lot_size', 'exchange', 'tick_size']

# Rearrange columns in each DataFrame
angelone_df = angelone_df[desired_order]
finvasia_df = finvasia_df[desired_order]
aliceblue_df = aliceblue_df[desired_order]
flattrade_df = flattrade_df[desired_order]
zerodha_df = zerodha_df[desired_order]
```

```python
# Remove rows with NaN values in all columns except 'tick_size'
angelone_df = angelone_df.dropna(subset=['token', 'name', 'trading_symbol', 'lot_size', 'exchange'])
finvasia_df = finvasia_df.dropna(subset=['token', 'name', 'trading_symbol', 'lot_size', 'exchange'])
aliceblue_df = aliceblue_df.dropna(subset=['token', 'name', 'trading_symbol', 'lot_size', 'exchange'])
flattrade_df = flattrade_df.dropna(subset=['token', 'name', 'trading_symbol', 'lot_size', 'exchange'])
zerodha_df = zerodha_df.dropna(subset=['token', 'name', 'trading_symbol', 'lot_size', 'exchange'])
```

```python
# Drop duplicate rows in each DataFrame
angelone_df.drop_duplicates(inplace=True)
finvasia_df.drop_duplicates(inplace=True)
aliceblue_df.drop_duplicates(inplace=True)
flattrade_df.drop_duplicates(inplace=True)
zerodha_df.drop_duplicates(inplace=True)
```

```python
print("Unique trading symbols in angelone_df:", angelone_df['trading_symbol'].unique())
print("Unique trading symbols in finvasia_df:", finvasia_df['trading_symbol'].unique())
print("Unique trading symbols in aliceblue_df:", aliceblue_df['trading_symbol'].unique())
print("Unique trading symbols in flattrade_df:", flattrade_df['trading_symbol'].unique())
print("Unique trading symbols in zerodha_df:", zerodha_df['trading_symbol'].unique())
```

```python
# Set trading_symbol as the index for each DataFrame
angelone_df.set_index('trading_symbol', inplace=True)
finvasia_df.set_index('trading_symbol', inplace=True)
aliceblue_df.set_index('trading_symbol', inplace=True)
flattrade_df.set_index('trading_symbol', inplace=True)
zerodha_df.set_index('trading_symbol', inplace=True)
```

```python
# Rename columns for clarity
angelone_df.rename(columns={
    'lot_size': 'angelone_lot_size',
    'token': 'angelone_token',
    'tick_size': 'angelone_tick_size',
    'exchange': 'angelone_exchange',
    'name': 'angelone_name'
}, inplace=True)

finvasia_df.rename(columns={
    'lot_size': 'finvasia_lot_size',
    'token': 'finvasia_token',
    'tick_size': 'finvasia_tick_size',
    'exchange': 'finvasia_exchange',
    'name': 'finvasia_name'
}, inplace=True)

aliceblue_df.rename(columns={
    'lot_size': 'aliceblue_lot_size',
    'token': 'aliceblue_token',
    'tick_size': 'aliceblue_tick_size',
    'exchange': 'aliceblue_exchange',
    'name': 'aliceblue_name'
}, inplace=True)

flattrade_df.rename(columns={
    'lot_size': 'flattrade_lot_size',
    'token': 'flattrade_token',
    'tick_size': 'flattrade_tick_size',
    'exchange': 'flattrade_exchange',
    'name': 'flattrade_name'
}, inplace=True)

zerodha_df.rename(columns={
    'lot_size': 'zerodha_lot_size',
    'token': 'zerodha_token',
    'tick_size': 'zerodha_tick_size',
    'exchange': 'zerodha_exchange',
    'name': 'zerodha_name'
}, inplace=True)
```

```python
# Start with the angelone_df
merged_df = angelone_df.copy()

# Join each DataFrame
merged_df = merged_df.join(finvasia_df, how='outer', lsuffix='_angelone', rsuffix='_finvasia')
merged_df = merged_df.join(aliceblue_df, how='outer', lsuffix='_finvasia', rsuffix='_aliceblue')
merged_df = merged_df.join(flattrade_df, how='outer', lsuffix='_aliceblue', rsuffix='_flattrade')
merged_df = merged_df.join(zerodha_df, how='outer', lsuffix='_flattrade', rsuffix='_zerodha')
```

```python
merged_df.to_csv('merged_data.csv')
```
